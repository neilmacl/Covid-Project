```python
## Replace all NaN with NA when stump or dead
mask = ((trees_subset['status'] == 'Dead') | (trees_subset['status'] == 'Stump'))
trees_subset.loc[mask] = trees_subset.loc[mask].fillna('NA')
trees_subset[(trees_subset['status'] == 'Dead') | (trees_subset['status'] == 'Stump')] 
```

```python
trees_subset.isnull().sum()
```

```python
## Replace nulls with most common answer
max_health = trees_subset['health'].value_counts().max()
max_spc_latin = trees_subset['spc_latin'].value_counts().max()
max_sidewalk = trees_subset['sidewalk'].value_counts().max()
max_problems = trees_subset['problems'].value_counts().max
```

```python
trees_subset['health'].fillna(max_health, inplace = True)
trees_subset['spc_latin'].fillna(max_spc_latin, inplace = True)
trees_subset['sidewalk'].fillna(max_sidewalk, inplace = True)
trees_subset['problems'].fillna(max_problems, inplace = True)
```

```python
trees_subset.isnull().sum()
```

```python
## Dealing with outliers
trees_subset.hist(bins=60, figsize=(20, 10))


```

```python
## Too big: stump_diam > 60, tree_dbh > 60 
big_trees = trees_subset[(trees_subset['stump_diam'] > 60) | (trees_subset['tree_dbh'] > 60)]
big_trees['status'].value_counts() 
```

```python
## Filtering out trees that seem too large and are likely input errors
small_trees = trees_subset[(trees_subset['stump_diam'] <= 60) & (trees_subset['tree_dbh'] <= 60)]
small_trees
```

```python
trees_subset['status'].value_counts()
```

```python
## Finding average tree size. Only inlcuding alive status or else condtions where tree is dead or stump would bias mean towards 0. Also filtering out values where tree_dbh = 0
small_trees_alive = small_trees[(small_trees['status'] == 'Alive') & (small_trees['tree_dbh'] != 0)]
small_trees_dead_or_stump = small_trees[(small_trees['status'] == 'Dead') | (small_trees['status'] == 'Stump')]
small_trees_alive.describe()
```

```python
small_trees_alive.shape
```

```python
stats_small_alive = small_trees_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_small_alive
```

```python
small_trees_alive = small_trees_alive.merge(stats_small_alive, on='spc_latin', how='left')
small_trees_alive
```

```python
## Making min tree_dbh for alive trees = 25th percentile
mask = small_trees_alive['tree_dbh'] < small_trees_alive['25%']
small_trees_alive.loc[mask, 'tree_dbh'] = small_trees_alive['25%']


```

```python
## Making max tree_dbh for alive trees = 75th percentile 
mask = small_trees_alive['tree_dbh'] > small_trees_alive['75%']
small_trees_alive.loc[mask, 'tree_dbh'] = small_trees_alive['75%']
```

```python
##small_trees_alive.describe()
small_trees_alive.shape
```

```python
##trees_subset.describe()
trees_subset_alive = trees_subset[trees_subset['status'] == 'Alive']
trees_subset_alive.shape
```

```python
trees_subset_alive = trees_subset_alive.merge(stats_small_alive, on='spc_latin', how='left')
trees_subset_alive
```

```python
## Fixing error where duplicates columns of 25% and 75% were added
trees_subset_alive = trees_subset_alive[['tree_id', 'tree_dbh', 'stump_diam',
       'curb_loc', 'status', 'health', 'spc_latin', 'steward',
       'sidewalk', 'problems', 'root_stone',
       'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
       'brch_light', 'brch_shoe', 'brch_other', '25%', '75%']]
```

```python
## Replacing outlier values in alive trees
mask = trees_subset_alive['tree_dbh'] < trees_subset_alive['25%']
trees_subset_alive.loc[mask, 'tree_dbh'] = trees_subset_alive['25%']

mask = trees_subset_alive['tree_dbh'] > trees_subset_alive['75%']
trees_subset_alive.loc[mask, 'tree_dbh'] = trees_subset_alive['75%']
```
